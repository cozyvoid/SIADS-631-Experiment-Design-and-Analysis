### Trust
- Trust plays an important role in economic and social transactions
    - Selling transactions (e.g., Ebay, Resnick et al. 2003)
    - Financial transactions (e.g., repay a loan)
    - Employment relationships (e.g., home office)
    - Friendship and family relations (e.g., not having an affair while married)
- Despite its importance, the measurement and definition of trust is not fully settled, and the identification of the exact role of trust in economic interactions has proven to be elusive
- Is trust just a side effect of good institutions or does it play an independent causal role capable of shaping important aggregate economic outcomes?

- Why is trust important?
    - Trust and economic growth
    - a way of reducing economic growth
    - Lubricating the economy
- Measuring trust:
    - Berg, Dickhaut and McCabe (1995): lab experiment
- Trust game as a predictor of economic behavior:
    - Karlan (2005)

### How can we measure Trust?

- Simply by asking people
- For example, **World Value Survey**
    - Generally speaking, would you say that most people can be trusted or that you can't be too careful in dealing with people?
    - Binary response format
        -  1. most people can be trusted, 2. need to be very careful
- How reliable are survey answers?
- It would be nice to have experimental tools that allow the measurement of trusting behavior
