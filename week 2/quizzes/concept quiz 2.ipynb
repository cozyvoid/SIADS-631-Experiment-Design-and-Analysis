### 1. What is the fundamental problem of causal inference? 

- Randomization does not result in similar treatment and control groups. 
- Correlation between two variables does not mean we can find causation. 
- **We cannot observe subjects in multiple experimental conditions at the same time.**
- Differences between conditions are not always statistically significant. 
- Treatments do not always have a reliable effect on dependent variables.

The correct answer is:

**We cannot observe subjects in multiple experimental conditions at the same time.**

That is the fundamental problem of causal inference: for any one unit or subject, we can observe the outcome under treatment or control, but not both at the same time. The unobserved outcome is the counterfactual

### 2. What is the average treatment effect? 
- Differences found between observations that are made of each subject before and after being treated. 
- **The difference between the average, or expected, outcomes of subjects in the treatment and control conditions.**
- A difference in outcomes between treatment and control conditions for each subject. 
- The average outcomes for each treatment in which subjects are randomly assigned to. 
- Hypothetical outcomes if subjects were treated or untreated.

The correct answer is:

**The difference between the average, or expected, outcomes of subjects in the treatment and control conditions.**

The average treatment effect (ATE) is the average difference in outcomes if everyone were treated versus if everyone were untreated.

### 3. A researcher plans to ask six subjects to donate time to an adult literacy program. Each subject will be asked to donate either 30 or 60 minutes. The researcher is considering two methods of randomizing the treatment. One method is to flip a coin before talking to each person and to ask for a 30-minute donation if the coin comes up heads or a 60-minute donation if it comes up tails. The other method is to write “30” and “60” on three playing cards each, and then shuffle the six cards. The first subject would be assigned the number on the first card, the second subject would be assigned the number on the second card, and so on. 

Say we increased the number of subjects from 6 to 600. What would happen to the balance in the number of subjects for both treatments? 

- For the coin flip, the balance would stay the same; for the playing cards, the balance would stay the same. 
- For the coin flip, it would be expected to be less balanced; for the playing cards, the balance would stay the same. 
- **For the coin flip, it would be expected to be more balanced; for the playing cards, the balance would stay the same.** 
- For the coin flip, it would be expected to be more balanced; for the playing cards, it would be expected to be less balanced. 
- For the coin flip, it would be expected to be more balanced; for the playing cards, it would be expected to be more balanced.

The correct answer is:

**For the coin flip, it would be expected to be more balanced; for the playing cards, the balance would stay the same.**

Why:

- With coin flips, treatment assignment is independent for each subject, so the proportions tend to get closer to 50/50 as the sample size grows.

- With the playing cards method, the design is exactly balanced by construction (equal numbers of 30 and 60), so increasing from 6 to 600 still keeps it balanced.

### 4. A researcher plans to ask six subjects to donate time to an adult literacy program. Each subject will be asked to donate either 30 or 60 minutes. The researcher is considering two methods of randomizing the treatment. One method is to flip a coin before talking to each person and to ask for a 30-minute donation if the coin comes up heads or a 60-minute donation if it comes up tails. The other method is to write “30” and “60” on three playing cards each, and then shuffle the six cards. The first subject would be assigned the number on the first card, the second subject would be assigned the number on the second card, and so on. 

## What is the expected value of the assigned number of minutes for the coin flip and card method respectively? 

- coin flip: 45; playing cards: 30 
- coin flip: 30; playing cards: 45 
- coin flip: 30; playing cards: 60 
- **coin flip: 45; playing cards: 45** 
- coin flip: 60; playing cards: 30

The correct answer is:

**coin flip: 45; playing cards: 45**

Why:

**Coin flip**: 30 or 60 minutes are each assigned with probability 0.5, so

$$
E[X] = 0.5(30) + 0.5(60) = 45
$$

**Playing cards**: there are three 30's and three 60's so the average assigned value is also

$$
\frac {3(30) + 3(60)} {6} = 45
$$ 

So both methods have the same expected value: 45 minutes.

### 5. Which of the following are examples of simple random assignment for assignment of 10 subjects to treatment and control groups? Select all that apply. 

- Placing subjects in descending height order and assigning the first five to the treatment and the rest to the control. 
- **Creating a random number generator that returns two numbers, 0 and 1, and having the generator produce 10 numbers while assigning subjects to a number. 0 represents the control and 1 represents the treatment.**
- Assigning each subject a random number from 0 to 100, and picking the subjects with the lowest five numbers for the treatment and the rest are for the control. 
- Placing 10 different colored pieces of paper in a container where five slips with one color represent the treatment and another five slips of a different color represent control, picking slips of paper out and handing them to subjects in a random order. 
- **Coin flip for each subject; heads receives the treatment and tails receives the control.**

So the correct answers are:

- **Creating a random number generator that returns two numbers, 0 and 1, and having the generator produce 10 numbers while assigning subjects to a number. 0 represents the control and 1 represents the treatment.**
- **Coin flip for each subject; heads receives the treatment and tails receives the control.**

Why the two you selected were marked wrong:

- Random number 0–100, then lowest five get treatment forces exactly 5 into treatment, so that is complete random assignment, not simple random assignment.
- Five treatment slips and five control slips drawn randomly also forces exactly 5 and 5, so that is complete random assignment too.

So the distinction is:

- Simple random assignment: independent random assignment for each subject, balance not guaranteed.
- Complete random assignment: random assignment with fixed group sizes.

### 6. Which of the following are examples of complete random assignment for assignment of 10 subjects to treatment and control groups? Select all that apply. 

- Placing subjects in descending height order and assigning the first five to the treatment and the rest to the control. 
- **Placing 10 different colored pieces of paper in a container where five slips with one color represent the treatment and another five slips of a different color represent control, picking slips of paper out and handing them to subjects in a random order.** 
- Creating a random number generator that returns two numbers, 0 and 1, and having the generator produce 10 numbers while assigning subjects to a number. 0 represents the control and 1 represents the treatment. 
- Coin flip for each subject; heads receives the treatment and tails receives the control. 
- **Assign each subject a random number from 0 to 100, picking the subjects with the lowest five numbers for the treatment and the rest are for the control.**

The correct choices are:

- **Placing 10 different colored pieces of paper in a container where five slips with one color represent the treatment and another five slips of a different color represent control, picking slips of paper out and handing them to subjects in a random order.**
- **Assign each subject a random number from 0 to 100, picking the subjects with the lowest five numbers for the treatment and the rest are for the control.**

Why:

- Complete random assignment means the number assigned to each group is fixed in advance here: exactly 5 treatment and 5 control.
- Both of the correct methods randomly assign subjects while guaranteeing a 5/5 split.

Why the others are not:

- Descending height order is not random.
- Random number generator returning 0 or 1 does not guarantee exactly 5 in each group.
- Coin flip for each subject also does not guarantee exactly 5 in each group.

### 7. In our sample of neighborhoods in a city, randomly assigned neighborhoods have optical fiber installed for internet connections. 

- Timing Randomization 
- **Access Randomization** 
- Encouragement Randomization

The correct answer is:

**Access Randomization**

Why:

- Access randomization means researchers randomly assign whether units get access to the treatment itself.
- Here, neighborhoods are randomly assigned to have optical fiber installed, which gives them direct access to better internet infrastructure.

For contrast:

- Timing randomization = everyone may get it, but the timing differs randomly.
- Encouragement randomization = researchers randomize encouragement to take up a treatment, not the treatment/access itself.

### 8. Match the following scenario to their randomization technique used: 

### A city announces optical fiber is being installed across the city for internet, however, only some neighborhoods can receive the installation in month 1, then the rest in month 2. We randomize those neighborhoods selected to have optical fiber installed first. 

- **Timing Randomization** 
- Access Randomization 
- Encouragement Randomization

The correct answer is:

**Timing Randomization**

Why:

- All neighborhoods will eventually receive optical fiber.
- What is randomized is when they receive it: month 1 versus month 2.

That is exactly *timing randomization*.

### 9. Match the following scenarios to their randomization technique used: 

### Optical fiber is installed across a city for internet, random neighborhoods in the city receive advertisements revealing the opportunity to activate usage of it. 

- **Encouragement Randomization** 
- Timing Randomization 
- Access Randomization

The correct answer is:

**Encouragement Randomization**

Why:

- The treatment opportunity already exists.
- What is randomized is the *encouragement* to use it, through advertisements.

So the neighborhoods are not being randomized into access or timing of installation, but into receiving an encouragement to activate the service.

### 11. Karlan (2005) applied the Trust Game in the context of microcredit lending. What attribute(s) does the decision of Player B (the recipient in the Trust Game) measure about Player B? 

- Risk attitude 
- Trust 
- Reliability 
- **Trustworthiness** 
- Bias

The correct answer is:

- **Trustworthiness**

In Karlan (2005), the standard interpretation of the Trust Game is that **Player A measures trust** and **Player B measures trustworthiness**. Karlan specifically reports that people identified as **trustworthy** in the game were less likely to default on microcredit loans.

### 12. What attribute(s) does the decision of Player A (the investor in the Trust Game) measure about Player A? Check all that apply. 

- Bias 
- Trustworthiness 
- **Trust** 
- **Risk attitude** 
- Reliability

The correct choices are:

- **Trust**
- **Risk attitude**

Why:

- In the Trust Game, **Player A’s decision** is usually interpreted as a measure of **trust**.

So for this question, select **Trust** and **Risk attitude**.
