### A Nationwide Tipping Experiment

- Uber randomized the launch of in-app tipping across 209 operational cities in the US and Canada in July 2017
    - On July 6, about half of the cities on Uber were given the option to tip. The feature was fully rolled out on July 17.
        - Can estimate short tun effects
    - Cities were assigned launch date by a randomized block design (matched pairs)
    - Cluster-randomized experiment with rich driver-level data on outcomes
    - ~ 815,000 drivers across 209 cities

### Experimental Design and Data

- 110 cities were assigned treatment by matched pairs
    - Stratified on whther city has UberPOOL and UberEats
    - Within strata formed matches by minimizing Euclidian distance across normalized city size, driver earnings, demographics, time since launch, and other variables
    - If a stratum had an odd number of cities the odd one out was randomly assigned to treatment or control
- Remaining cities randomlu assigned to treatment or control
- 105 control cities and 104 treated cities

### Randomization (Balanced) Check

- Administrative errors are possible whenever one generates primary data
- Experimenters bear the burden of demonstrating the soundess of their experimental procedures
- When reporting experimental results, researchers should give a deatiled description of the random assignment process and any instance in which subjects are lost or excluded after allocation

### Overall Market Effects

- We do not have statistically significant changes for any of these metrics, but based on the point estimates we see a short-run
    - 0.4% ($\pm$ 2.7%) changes in minutes worked
    - -1.3% ($\pm$ 5.5%) change in rider sessions
    - -0.75% ($\pm$ 4.0%) change in utilization
    - -1.4% ($\pm$ 6.7%) change in total hourly earnings
    - -1.1% ($\pm$ 6.5%) change in organic hourly earnings
    - -0.7% ($\pm$ 6.9%) change in post-commission hourly earnings (Uber does not collect commissions on tips)
- We also observe a statistically insignifant drop in surge earnings
- An interpretation (?): Increase in labor supply, decrease in deman leads to a fall in utilization and surge that results in a fall earnings

### Dynarski: Target Encouragement for Admission to UM

- Combining encouragement to apply to an selective university with the promise of free tuition
- Low-income students generally unlikely to attend selective colleges
- Targeted contact through sending postcards to such students
- Encouraged students to apply with the promise of free tuition
- Found that students were:
    - More than twice as likely to apply
    - And more than twice as likely to enroll
- Those who enroll are made up of students who would:
    - Attend a less-selective college
    - A community college
    - No college
