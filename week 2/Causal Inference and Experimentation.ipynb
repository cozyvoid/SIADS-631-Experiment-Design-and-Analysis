### Notation

- Indexing experimental subjects: In our system of notation, the subscript *i* refers to subjects 1 through *N*.
- Treatment: The variable *d_i* indicates whether the *i*th subject is treated 
    - binary treatments:
        - $ d_i $ = 1 means the *i*th subject receives the treatment, 
        - $ d_i $ = 0 means the *i*th subject does not receive the treatment. It is assumed that $ d-i $ is observed for every subject.
- Comment: In practice, you must define what it means to be treated and measure whether each subject is treated.

### Potential Outcomes

- Regardless of which treatment a subject receives, that subject may be said to have a potential response in the event that treatment is or is not received.
- **Potential outcomes** may be written $ Y_i(d) $, where the arguemtn *d* indexes the treatment.
- These potential outcomes are fixed attributes of each subject and represent the outcome that would be observes hypothetically if that subjet were treated or untreated.
- **Binary treatments**:
    - $ Y_i(1) $: potential outcome if the *i*th subject were treated.
    - $ Y_i(0) $: potential outcome if the *i*th subject were not treated.

### Random assignment and unbiased inference

- Fundamental problem of causal inference (Holland 1986): we observe only onu potential outcome
- this problem seems more fundamental in some problems than others
- For example, we could assign some subjects to be asked "On a scale of 1-10, where 10 is the highest, tell me how much you like vanilla ice cream?" Other subjects could be assigned to the same question, but substituting choclate ice cream.
- In that design, we observe either $ Y_i (vanilla) $ or $ Y_i (chocolate) $. But why not ask subjects both questions?
- **Answer**: One might be concerned about order effect.

### Observed Outcomes

- Observed outcomes: The connected between the observed outcome $ Y_i $ and the underlying potential outcomes is given by the equation

$$ Y_i = d_iY_i(1) + (1 - d_i) Y_i (0) $$

- This equation indicates that the $ Y_i (1)  $ are observed for subjects who are treated, and the $ Y_i (0) $ are observed for subjects who are not treated.
- For any subject, we observe either $ Y_i (1) $ or $ Y_i (0) $, not both
    - Smoking example:
        - "treated": smoker
        - "control": non-smoker

### Radnom assignment and the "missing data" problem

- Random assignment addresses the "missing data" problem by creating two groups of observations that are, in expectation, identical prior to application of the treatment 
- Treatment and control groups have same expected outcomes:

$$
E[Y_i(1)|D_i = 1] = E[Y_i(1)]
$$
$$
E[Y_i(1)|D_i = 0] = E[Y_i(1)]
$$
$$
E[Y_i(1)|D_i = 1] = E[Y_i(1)|D_i = 0]
$$

- Example: random assignment used in rural India to assign women in head 1/3 of local village councils

### Random assignment

- **Random assignment** refers to a procedure that allocates treatments with known probabilies that are greater than zero and less than one.
- The most basic forms of random assignment allocate treatments such that **every subjects has the same probability of being treated**.
- Let *N* be the number of subjects, and let *m* be the number of subjects who are assigned to the treatment group. Assume that *N* and *m* are integers such that *0 < m < N*.

### Simple vs. Complete Random Assignments

- Simple random assignment refers to a procedure whereby each subjet is allocated to the treatment group with probability $ \frac {m} {N} $
- Complete random assignment refers to a procedure that allocates exactly *m* units to treatment
- Number of possible allocations of subjects to treatment or control conditions under complete random assignment:

$$

\frac {N!} {m! (N - m)!}

$$

> Recall: *N* is the size of the sample and *m* is the number of subjects in the treatment



### *Difference-in-means estimator (DIM) under complete random assignment*

- Suppose the $ D_i $ were randomly assigned such that *m* subjects were selected for treatment and the remaining *N - m* to control.
- Equivalent to sorting the observations randomly and calling the first *m* subjects the treatment group.
- Assume that subjects have been ordered in this random fashion. The expected difference-in-means estimator (DIM) is equal to the ATE.
