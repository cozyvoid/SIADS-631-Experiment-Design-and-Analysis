## Wei Ai, Roy Chen, Yan Chen, Qiaozhu Mei, Webb Phillips
### Proceedings of the National Academy of Sciences (PNAS) 2016

### The Problem

- Three billion people subsisting on $2.5 per day
- Micro and small enterprises: the largest employers in many developing countries
    - Growth stifled by a lack of access to credit or other financial services
- Microfinance programs provide small loans and other financial services
    - Online peer-to-peer microlending



- "Lend as little as $25 to create opportunity for people around the world." (zero interest)
- $1.3 billion loans funded through Kiva
- 3.2 million borrowers in more than 80 countries
- 1.8 million lenders

### Kiva's Challenge



- Lending Frequency
    - Few lenders made many loans;
    - Many lenders made few loans;
    - One-third of users never made a loan
---
- How do we increase lender participation?
- Solution: team competition

### Team Competition: Leaderboards




### Does joining a team increase lending?



### Hypotheses

- Lenders will be more likely to join teams if we make "good" recommendations

- What's a "good" recommendation?
    - Location similarityL homophily (Kaggle competition)
    - Loan history similarity: homophily
    - Leaderboard positions: status

- Users will lend more after they join teams

### Experiment Design: Sample Selection

- Field experiment: May 2014
- Sample selection criteria (from Kiva):
    - Haven't joined any team
    - Have location information in their profile
    - Allow marketing email, set their pages public
    - have made at least 2 laons in the past 6 months
- 69,845 users met these criteria

### Experimental Design: 3 x 2 factorial

### Team Recommendation" Emails

### Treatment Effect on Joining Teams

- Every treatment except "Teams-Exist" did significantly better than the control.
- **Location with explanation** has the largest effect.
- Among those who opened email,  two treatments did significantly better than the team-exist treatment:
    - **Location-explanation (survives MHT correction)**
    - History-explanation (insignificant after MHT correction)

### Effect of Team Membership on Lending Amount



- First stage (**inclusion restriction**): F-statistics = 23.55; strong instrument;
- Second stage (**exclusion restriction**): Email itself does not increase lending.



### Nuts and Bolts: Three Important Details about IV

1. You can use multiple instrumental variables at the same time
2. IV does not give you the ATE
    - IV methods identify the average gains to persons induced to change their choice by a chance of the instrument (compliers)
        1. ... only for the group influenced by the treatment ("localy average treatment effect" or **LATE**)
        2. Caution in extrapolating to the whole population
3. You must account for estimation error in the first stage when computing standard errors in the second stage.

### Bottom Line

- IV is only as good as the instruments on which it is based
    - **Inclusion restriction**: weak instrument don't work well (well F-test to assess objectively)
    - **Exclusion restriction**: It is difficult to test empirically
        - Exception: if you have more than one plausible instrument, you can see if they both give you the same answers (Sargan test or Wu-Hausman test)

## Summary
- Team recommendation emails significantly increase the likelihood a lender joins a team compared to the control
    - **Location + explanation** has the largest
- For those who joined a team, their average lending amount increased by
    - **$392 in the 7-day window**
- Support team membership as an effective mechanism for promoting pro-social behavior
- Method:
    - Anticipated one-sides non-compliance
    - Built an instrument(*ex ante*) into the experimental design
