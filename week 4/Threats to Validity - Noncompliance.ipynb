### Noncompliance

- **One-sided noncompliance** or **failure-to-treat** occurs when some of the subjects assigned to the treatment group do not actually receive the treatment
- **Two-sided noncompliance** occurs when some subjects in the assigned treatment group go untreated and some subjects in the assigned control group receive the treatment

### Assigned vs. Actual Treatment

- Let the experimental assignment of subject *i* be denoted **z_i**
    - When $z_i = 1$, the subject is assigned to the treatment group
    - when $z_i = 0$, the subject is assigned to the control group.
- In previous lectures, $d_i = z_i$:
    - all subjects assigned to the treatment group are treated, and
    - all subjects assigned to the control group are treated.
- Let $d_i(z)$ represent whether subject *i* is actually treated when threatment assignemtns is *z*.
    - if a subject would be treated if assigned to the treatment group, $d_i(z=1) = d_i (1) - 1$.
    - If a subject would receive no treatment if assigned to the treatment group, $d_i(1) = 0$.

### Compliers and Never-Takers

- Subjects in experiments with one-sided noncompliance can be divided into two (latent) groups.
- **Compliers** meet two conditions:
    - they recevie the treatment if assigned to the treatment group $(d_i(1)=1)$, and 
    - they do not receive the treatment if assigned to the control group $(d_i(0)=0)$.
- **Never-Takers** never take the treatment regardless of whether they are assigned to the treatment group or control group
    - $d_i(1)=0$ and $d_i(0) = 0$

### Estimating ATE with noncompliance

- One-side noncompliance
    - Treatment group
        - Compliers
        - Never-takers
- Wrong approaches
    - ignore compliance issue
    - Compare compliers to the control condition: bias

### Optimal Design Under Low Rates of Compliance

- Deploy two versions of the treatment, one with the active ingredient and another that functions merely as a **placebo**.
- A comparison of those who actually receive the treatment/placebo is a comparison of treated and untreated Compliers
- Ai *et al.* (2016)
    - Receive email which inform them of the existence of lending teams (placebo)
    - Or receive team recommendations (treatment)

###  Spillovers

- Spillovers is an indirect effect of a program on subjects who are not being treated by the experiment.

### Hawthorne Effect

- When the treatment group is aware of being selected and increases their productivity beacause they feel special
- The "Illumination Experiment" at Western Electric's Hawthorne Plant in the 1920's

### John Henry Effect

- A control is aware of their status as members of the control group and is able to compare their performance with that of the treatment group
- May actively work harder to overcome the "disadvantage" of being in the control group
- John Henry, "a legendary American steel driver in the 1870's who, when he heard his output ws being compared with that of a steam drill, worked so hard to outperform the machine he died in the process." -- Saretsky (1972)

### Experimenter Demand Effect 

- When participants change their behavior to fit what they believe the experimenter is trying to test.
- The Stanford prison experiment (1971) (by Dr. Philip G. Zombardo) had college students volunteer to play guards or prisoners; both took their roles too far and the experiment was cut short after only 6 days.
- A demonstration rather than an experiment

### Anticipation Effect

- When the control group change their behavior because they think they will receive the program in the future.

### Survey Effects

- When being survery can change future behavior
