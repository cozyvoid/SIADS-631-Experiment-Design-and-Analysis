# Threhold Public Good

## Overview

- each round, each group member received same amount of money
- each simulataneously chooses how much to contribute to water purification project
- project proceeds only if **total contribution at least as large as project's cost (threshold)**
    - **Refund** if theshold is not reached
- **each** individual's benefit from project

$ project cost x rate of return $

### Your payoffs

Overview
- depend on yours and total contributions
- excess contributions (relative to threshold): lost
- insufficient contributions: **refund**

An example
- start with $20; $50 project with 30% rate of return
- you contributed $12 out of $50 contributed by group



- All players are endowed with 20 ECU for each round
- Group for 5, with 4 other bots
- Bot strategy:
    - fair share to reach the threhold + random number between -3 and 3
    - "Fair share"
        = threshold/group site
        = 50/5
        = 10



### What have we learned from the experiment?

- Theory makes precise predictions, assuming everyone is selfish
    - A set of inefficient Nash equilibria: $x_i = 0 $
    - A set of efficient Nash equilibriaL contributing you "*fair share*" so the sum of contributions each the threshold
- What have we learned from the data?
    - Most students contribute theri fair share in round 1: $ x_i = 10 $
    - they learned to contribute a littel more because of the noisy robot strategies
- why do we include this experiment?
    - To illustrate how the threshold changes the incentive to free ride
    - Kiva (next module) used a threshold public good model to provide laons to low income entrepreneurs in developing countries
