# Concept Quiz 4

### 1. In a field experiment, you observe that some subjects in a treatment group are untreated, whereas some subjects in the control group receive the treatment. What do we call this phenomenon?

- **Two-sided noncompliance**
- One-sided noncompliance
- Attrition
- Spillovers

The correct answer is **Two-sided noncompliance**.

That is when:

- some people assigned to treatment do not actually receive it, and
- some people assigned to control do receive it.

For contrast:

- One-sided noncompliance: only one group fails to follow assignment, usually treatment assignees not taking treatment while control assignees cannot access it.
- Attrition: subjects drop out or outcomes are missing.
- Spillovers: one subject’s treatment affects another subject’s outcome.

### 2. In Ai et al. (2016), when the experimenters sent out emails to Kiva lenders, only one-third of the lenders actually opened their emails. What do we call this phenomenon?

- Two-sided noncompliance
- **One-sided noncompliance**
- John Henry Effect
- Attrition

The correct answer is **One-sided noncompliance**.

Here, the treatment was **being sent/opening the email encouragement**, but only some of those assigned to receive it actually engaged with it. The control group was not described as receiving the treatment, so noncompliance is only on one side.

### 3. Subjects in experiments with one-sided noncompliance can be divided into two (latent) groups, compliers and never-takers. Which conditions must compliers satisfy? Check all that apply.

- **They receive the treatment if they are assigned to the treatment group.**
- **They do not receive the treatment if they are assigned to the control group.**
- They do not take the treatment if they are assigned to the treatment group.
- They receive the treatment if they are assigned to the control group.

The correct choices are:

- **They receive the treatment if they are assigned to the treatment group.**
- **They do not receive the treatment if they are assigned to the control group.**

Compliers are the subjects who follow their assignment. In a one-sided noncompliance setup:

- if assigned to treatment, they take the treatment
- if assigned to control, they do not receive it

### 4. You are conducting an experiment consisting of a treatment and a control condition. If you encounter one-sided noncompliance, which of the following statements are true?

- **Among the treated subjects, you can separate the compliers from the never-takers.**
- Among control subjects, you can separate the compliers from the never-takers.
- **Among control subjects, you cannot differentiate between the compliers and the never-takers.**
- You will never know which of your subjects are compliers and which of your subjects are never-takers.

The true statements are:

- **Among the treated subjects, you can separate the compliers from the never-takers.**
- **Among control subjects, you cannot differentiate between the compliers and the never-takers.**

Why:

- In one-sided noncompliance, only those assigned to treatment can fail to take it.
- So in the treatment group, you can observe who actually took treatment (compliers) and who did not (never-takers).
- In the control group, both compliers and never-takers appear untreated, so you cannot tell them apart.

### 5. After reading the following scenario, decide which evaluation-driven effect is occurring:

### *A company plans to roll-out its low-cost delivery meal-kits to 50 metropolitan areas. To assess the effect of their meal-kits on eating at restaurants, they are releasing the meal-kits to these metropolitan areas in waves, randomly selecting 10 metropolitan areas each month to release the meal-kits to until all 50 have access. However, in preparation of receiving access to the meal-kits, many consumers in these areas begin changing their habits and go to restaurants more.*

- Survey Effect
- **Anticipation Effects**
- Hawthorne Effect
- John Henry Effect

The correct answer is **Anticipation Effects**.

This is happening because people change their behavior **before they actually receive the treatment**, since they know it is coming. Here, consumers start changing restaurant-going habits in preparation for the meal-kit rollout.

### 6. After reading the following scenario, decide which evaluation-driven effect is occurring:

### *A factory instituted a 100% automation policy where processes are expected to run constantly without human oversight except to repair major failures. The factory manager decided to test if instituting process managers would reduce the failure rate of factory processes and decrease costs of downtime. The factory manager hired process managers and randomly assigned them to 25 of the 50 processes currently running at the factory. The factory manager directly monitored process managers for a month as part of the experiment. After the study, the factory manager found an 80% decrease in the failure rate of monitored processes compared to unmonitored processes. However, since the factory manager stopped monitoring the process managers, it was found there was only a 15% decrease in the failure rate.*

- John Henry Effect
- Anticipation Effect
- Experimenter Demand Effect
- **Hawthorne Effect**

The correct answer is **Hawthorne Effect**.

The key clue is that performance improved a lot **while subjects were being monitored**, then dropped once the monitoring stopped. That is the **Hawthorne Effect: people change their behavior because they know they are being observed.**

### 7. Using instrumental variables, what are the problems we try to correct for?

- **Endogenous decision to participate**
- **Error terms correlate with independent variables**
- Omitted subjects
- **Omitted variable bias**

The correct answers are:

- **Endogenous decision to participate**
- **Error terms correlate with independent variables**
- **Omitted variable bias**

Why:
Instrumental variables are used when treatment or participation is endogenous, meaning it is related to unobserved factors in the error term. That also means the independent variable is correlated with the error term, often because of omitted variable bias.

Not correct:

- Omitted subjects is not the standard problem IV is designed to fix.

### 8. *We wish to evaluate a voluntary job training program by comparing outcomes for those who participate to outcomes of those who do not participate. A social worker visits unemployed people to encourage them to participate in this program. She only visits 50% of persons on her roster, and she randomly chooses whom she will visit. If she is effective, many people she visits will enroll, i.e., there will be a correlation between receiving a visit and enrolling. However, visit does not have direct effect on outcomes (e.g. income) apart from its effect through enrollment in the training program.*

### Which of the following variables is a good candidate for an instrumental variable in this setting?

- Participation decision
- Income
- Opportunity cost of participation
- **Receiving a visit**

The correct answer is **Receiving a visit**.

Why it works as an instrument:

- it is randomly assigned
- it is correlated with enrollment in the training program
- it affects income only through participation, not directly

That matches the core requirements for a valid instrumental variable.

### 9. An instrumental variable is valid if it satisfies the following conditions:

- **Exclusion restriction**
- **Inclusion restriction**
- Independence of the endogenous regressor
- Correlation with the error term

Based on the grading shown, the correct choices are:

- **Exclusion restriction**
- **Inclusion restriction**

Why your earlier choice missed:

- Independence of the endogenous regressor is not a validity condition for the instrument. The issue is actually that the instrument should be independent of the error term, not the endogenous regressor.
- Correlation with the error term is not valid; a good instrument must not be correlated with the error term.

### 10. The Instrumental Variable analysis is only as good as the instruments on which it is based. Evaluate if the following statement is true or false:

### *You can use an F-test to assess objectively if an instrument satisfies the inclusion restriction.*

- **True**

True

An F-test from the first-stage regression is commonly used to assess whether the instrument is sufficiently correlated with the endogenous variable, which is the inclusion restriction (also called instrument relevance). A very low first-stage F-statistic suggests a weak instrument.

### 11. The Instrumental Variable analysis is only as good as the instruments on which it is based. Evaluate if the following statement is true or false:

### *There is no objective measure to assess whether an instrument is strong or weak.*

- **False**

False

There is an objective way to assess instrument strength, typically using the first-stage F-statistic. A low F-statistic suggests a weak instrument.
