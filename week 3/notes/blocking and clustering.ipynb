# Block and Clustering

### Randomizing: Block and Clustering

Two types of randomization
- Blocking (stratifying) can increase power
- Clustering can decrease power

## Block (Stratified) Random Assignment

**Block Random Assignment**: A procedure whereby subjects are partitioned into subgroups (called **blocks** or **strata**), and complete random assignment occurs within each block.

Example: suppose we have 20 subjects in our experiment, 10 men and 10 women.

- Suppose our experimental design calls for 10 subjects to be placed into the treatment condition.
- Block random assignment ensures equal numbers of men and women will be assigned to each experimental condition.
- First, we partition the subject pool into men and women. 
- From the pool of male subjects, we randomly assign five into the treatment group; from the poor of female subjects, we randomly assign five into the treatment group.

### Advantages of Blocking

- **Administative requirements** sometimes specify how many subjects of a given profile must be in the treatment group, a requirement that blocking can easily accommodate;
- Block designed facilitate **subgroup analysis** by ensuring that (blocked) subgroups are avilable for study;
- Blocking on prognistic variables may **improve the precision** with which the ATE is estimated

## Clustered Random Assignment

- **Clustered assignmen**t**: all subjects in the samee cluster are placed into either the treatment or control condition
- Clustered assignment is sometiems a **design compromise** occasioned by technology or the practrical constraints of administering a treatment
    - **Example: Uber tipping experiment**

### Background

- Prior to the summer of 2017 UBer did not support in-app tipping
- Uber launched optional in-app as its first stage in its "180 Days of Change" campagin
    - One of the more resquested product changes by riders and drivers in feedback to the company
    - Launched as an experiment across different cities
- How does tipping affect:
    - Labor Supply?
    - Demand?
    - Earnings?
    - Quality of Service?

### A Nationwide Tipping Experiment

- Uber randomized the launch of in-app tipping across 209 operational cities in the US and Canada in July 2017
    - On July 6, about half of the cities on Uber were given the option to tip. The feature was fully rolled our on July 17th.
        - Can estimate short run effects
    - Cities were assigned launch date by a randomized block design (matched pairs)
    - Cluster-randomized experiment with rich driver-level data on outcomes
    - ~815,000 drivers across 209 cities

### Experiment Design and Data

- 110 cities were assigned treatment by matched pairs
    - Stratified on whether city has UberPOOL and UberEats
    - Within strata formed matches by minimizing Euclidian distance across normalized city size, driver earnings, demographics, time since launch, and other viariables
    - If a stratum had an odd number of cities the odd one out was randomly assigned to treatment or control
- Remaining cities randomly assigned to treatment or control
- 105 control cities and 104 treated cities

### Clustering: An Example

**Cluster random assignment (at school level)**: the experimenter randomly assigns two of the four schools of the treatment group   
    - the total number of possioble random assignments is: $ \frac {4!}{2!2!} = 6 $
    - Across all randomization, the true ATE = 4
    - Calculate the standard deviation of the estimated ATE across all possible random assignments: **SE = 2.9**

**Complete random assignment (at classroom level)**: allocated the 12 classrooms, half to treatment and half to control
    - the number of possible randomizations is $ \frac {12!}{6!6!} = 924 $, 
    - and across them the standard error is **SE = 1.6**
Clustering led to an **81% increase in the standard error**, which is roughly the same increase that would have occurred if our sample size had declined from 12 to 4.

### Clustered Random Assignment: Implications for estimation

- Clustered assignment affects the sampling distribution. In most cases, clustering will cause subjects with similar potential outcomes to be assigned together to the same experimental group, which increases sampling variabiliity.
- Second, clustered assignment with **clusters of unequal size** (i.e., different numbers of fsubjects per cluster) presents a problems for difference-in-means estimation; this estimator may not be unbiased when the number of cluseters is small. The size and magnitude of the bias depend on how potential outcomes covary with cluster size. This bias goes away as the number of clusters increases.
