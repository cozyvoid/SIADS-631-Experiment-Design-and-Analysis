### The Boy Who Cried Wolf: Type I vs. Type II error 

- **Null Hypothesis**: There is no wolf.
- **Alternate Hypothesis**: There is a wolf.

- **Type I Error**: Villagers believed the boy when there was no wolf. (Rejecting null hypthesis incorrectly)
- **Type II Error**: Villagers did not believe the boy when there actually was a wolf (Rejecting alternate hypothesis incorrectly)

### Power of the Design 

*"The power of the design is the probability that, for a given effect size and a given statistical significant level, we will be able to reject the hypothesis of zero effect."*

- Duflo, E., Glennerster, R. & Kremer, M. (2007)

(what's the **probability of detecting a real effect**?)

### Approach to sample size determination

1. Determine the expected difference (or minimum average treatment effect)
2. Find out the standard deviations of both groups (from prior empirical studies, pilot, or theory)
3. Set alpha error to be tolerated (vis. type I error, P = 0.05)
4. Decide the power of the study desired (vis. 80%, type II error 0.2)
5. Select the appropriate formula 
6. Calculate the sample size using the formula
7. Give allowance for drop-out rate
8. Give allowance for non-compliance of treatment if possible
