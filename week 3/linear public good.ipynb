- Each simultaneously chooses how much to contribute to a **public good** (water purification project)
- **Each** individual's benefit from project:
    - total group contributions x rate of return

### Provisions of Public Goods

- Government: through taxes
- Private Provision of Public Goods
    - **Why do people donate to public goods** 
        - Social reward (reputation, etc.)
        - Social punishment
        - Incremental increase of public goods worth that much
        - Altruism
        - Positive emotions: interests, fun
    - **Free-riding**: choosing not to donate to a public good but still benefitting from the donation of others
    - **Social loafing**: exert less effort in team work
    

### Voluntary Contribution to Public Goods: An Example

- Each participant has 20 tokens in each round
- Tokens may be allocated to a private account or group account
    - Private account: converted to cas at a constant rate
    - Group account:
        - Yield a lower return to the individual (0.5)
        - Yield an additional return that accrues to each participant in the group, including the contributer (0.5)

### Analysis of linear public good experiment: n = 2

- Suppose player 1 contributes $x_1$ tokens into group account
    and player 2 contributes $x_2$ tokens into group account

- Rate of return from group account = 0.5
- individual payoffs:
$$
\pi_1 = 20 - x_1 + 0.5 (x_1 + x_2) = 20 - 0.5x_1 + 0.5x_2 
$$
$$
\pi_2 = 20 - x_2 + 0.5 (x_1 + x_2) = 20 - 0.5x_2 + 0.5x_1
$$

### Analysis of linear public good experiment: n > 2

- Dominant strategy is to contribute nothing to group account: $ x_1 = x_2 = 0 $
- With $n$ players



- Dominant strategy is still to contribute nothing to group account: $x_i = 0$
    - However, when $n>2$, the Pareto optimal outcome is to contribute everything
        - Agent $i$ loses $0.5x$, but others gain $0.5nx$
        - Total social value is greater as $x_i$ increases and each individual is better off, too 



- All players are endowed with 20 ECU for each round

    - In subsequent round, bots act as conditional cooperators: They contribute the average contributions of the other three players from the previous period

### What have we learned from the experiment?

- Theory makes precise predictions, assuming everyone is selfish
    - Dominant strategy: $x_i = 0$, contributing nothing
- What have we learned from the data?
    - More students contribute zero in the last three rounds
    - Some students contribute 20 ECU
    - Most students are conditional cooperators
- Why do we include this experiment?
    - To illustrate the incentives to free-ride on others' contributions when a public good is present
