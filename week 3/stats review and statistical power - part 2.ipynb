### a. 0/1 treatment,  equal outcome variances 

### Framework and Notations

- **Treatment**: $T = 1$; Control: $T = 0$
- Assume outcome $X_0$ is $N(\mu_0, \sigma^2_0)$ and $X_1$ is $N(\mu_1, \sigma^2_1)$
- The **minimum detectable effect**: $\sigma_1 - \sigma_0 = \delta$
- **Null Hypothesis** $H_0: \sigma_0 = \sigma_1$
- **Alternate Hypthesis** $H_1: \sigma_1 - \sigma_0 = \delta > 0$
- We need the **difference in sample means** $X_1 - X_0$ to statisy:
    1. **Significance level** (probability of Type I error) = $\alpha$
    2. **Power** (1 - probability of Type II error) = $1 - \beta$
