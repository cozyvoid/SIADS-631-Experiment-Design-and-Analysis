# Concept Quiz 3

**1. What is a Type I Error?**

- Accepting the null hypothesis when it was correct.
- Accepting the alternative hypothesis when it was correct.
- **Rejecting the null hypothesis when it was correct.**
- Rejecting the alternative hypothesis when it was correct.

A Type I error is a false positive: concluding there is an effect when there actually is not.

**2. What is a Type II Error?**

- Accepting the null hypothesis when it was correct.
- **Rejecting the alternative hypothesis when it was correct.**
- Rejecting the null hypothesis when it was correct.
- Accepting the alternative hypothesis when it was correct.

A Type II error is a false negative: failing to detect a real effect.

**3. What is power (of a test) equivalent to?**

- probability of Type I error
- **1 - probability of Type II error**
- probability of Type II error
- 1 - probability of Type I error
- 1 - probability of a true alternative hypothesis

Power is the probability of correctly rejecting the null hypothesis when the alternative is true, which equals 1 - β, where β is the probability of a Type II error.

**4. Which of the following are true regarding their effect on power? Select all that apply.**

- [x] A larger sample size increases power
- [x] A larger effect size increases power
- [_] larger standard deviation increases power
- [x] A larger standard deviation decreases power
- [_] A smaller effect size increases power
- [_] A smaller sample size increases power

Power goes **up** when it is easier to detect a real effect. Larger samples and larger effects make detection easier, while more variability (larger standard deviation) makes detection harder.

**12. Which of the following are true regarding the difference-in-differences estimator?**

- [x] **The difference-in-differences estimator is unbiased when treatments are randomly assigned.**
- [x] **The differencing operation reduces variance in treated and untreated outcomes.**
- [_] Only outcomes from after the experiment are measured whereas difference-in-means compares outcomes from before and after the experiment.
- [_] It is usually less precise than the difference-in-means estimator.

These are true:

- The difference-in-differences estimator is unbiased when treatments are randomly assigned.
- The differencing operation reduces variance in treated and untreated outcomes.

The other two are false. Difference-in-differences uses **before and after** outcomes, and it is often **more precise**, not less precise, than a simple difference-in-means estimator when pre-treatment and post-treatment outcomes are correlated.
